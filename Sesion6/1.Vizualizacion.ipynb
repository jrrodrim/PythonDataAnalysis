```python
#Graficos de linea- Mostrar relaciones en variables continuas

#Datoss
x=[1,2,3,4,5]
y=[2,4,5,7,8]
```

```python
#Creacion del grafico

import matplotlib.pyplot as plt

plt.plot(x,y)

plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Grafico de lineas")

```

```python
#Grafico de dispersion

plt.scatter(x,y)
plt.show
```

```python
#Histograma
#Libreria para generar ddatos aleatorios NUmpy
import numpy as np
datos = np.random.rand(1000)
plt.hist(datos,bins=30, color='skyblue',edgecolor= 'black')
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show

```

```python
#GRafico de barros, comparr categorias o grupos

categorias = ['a','b','c','d']
valor= [30,40,50,40]


plt.bar(categorias, valor, color='green')
plt.show
```
